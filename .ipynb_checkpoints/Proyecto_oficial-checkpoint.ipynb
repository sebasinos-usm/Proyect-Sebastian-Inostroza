## FASE 1 - ESTABLECER DATASET
El datase esta compuesto de una carpeta raiz llamada "informes_pdf" que contiene 2 sub carpetas de archivos en PDF de informes Criticos y No Criticos.
El codigo tiene la finalidad de cambiar el formato de PDF a TXT, considerando repetar formatos y manejando posibles errores. 
La salida es una carpeta raiz con las mismas subcarpetas pero con archivos en TXT.

```python
import fitz  # PyMuPDF
import os

def extraer_texto_de_pdf(ruta_pdf):
    """
    Extrae el texto de un archivo PDF preservando el formato y la disposici√≥n
    de los bloques de texto.
    """
    try:
        doc = fitz.open(ruta_pdf)
        texto_completo = ""
        for pagina in doc:
            # 1. Extraer texto en 'bloques' para preservar el layout.
            #    El argumento sort=True ordena los bloques de arriba a abajo.
            bloques = pagina.get_text("blocks", sort=True)
            
            # 2. Iterar sobre cada bloque y a√±adir su texto.
            for b in bloques:
                # El √≠ndice 4 de la tupla del bloque contiene el string de texto.
                texto_del_bloque = b[4]
                texto_completo += texto_del_bloque
            
            # Agregamos un separador claro entre p√°ginas
            texto_completo += "\n--- Fin de P√°gina ---\n"
            
        doc.close()
        return texto_completo
    except Exception as e:
        print(f"  - ERROR al leer el archivo {os.path.basename(ruta_pdf)}: {e}")
        return ""


def procesar_pdfs_recursivamente(carpeta_raiz_entrada, carpeta_raiz_salida):
    """
    Convierte a texto todos los PDFs de una carpeta y sus subcarpetas,
    replicando la estructura en una carpeta de salida.
    """
    print(f"Iniciando conversi√≥n recursiva desde: '{carpeta_raiz_entrada}'")
    archivos_procesados = 0

    for dirpath, _, filenames in os.walk(carpeta_raiz_entrada):
        
        ruta_salida_actual = dirpath.replace(carpeta_raiz_entrada, carpeta_raiz_salida, 1)
        if not os.path.exists(ruta_salida_actual):
            os.makedirs(ruta_salida_actual)
            print(f"Creando directorio: {ruta_salida_actual}")

        for nombre_archivo in filenames:
            if nombre_archivo.lower().endswith(".pdf"):
                
                ruta_pdf_completa = os.path.join(dirpath, nombre_archivo)
                
                nombre_base = os.path.splitext(nombre_archivo)[0]
                nombre_txt = nombre_base + ".txt"
                ruta_txt_completa = os.path.join(ruta_salida_actual, nombre_txt)
                
                print(f"  - Convirtiendo: {nombre_archivo}")

                texto_extraido = extraer_texto_de_pdf(ruta_pdf_completa)
                
                if texto_extraido:
                    with open(ruta_txt_completa, "w", encoding="utf-8") as archivo_salida:
                        archivo_salida.write(texto_extraido)
                    archivos_procesados += 1

    print(f"\n‚úÖ ¬°Proceso completado! Se convirtieron {archivos_procesados} archivos PDF a texto.")
    print(f"Los resultados se guardaron en: '{carpeta_raiz_salida}'")

# --- USO ---

CARPETA_RAIZ_PDFS = 'Informes_pdf'
CARPETA_RAIZ_TEXTOS = 'Informes_en_Texto'

# Creamos carpetas de ejemplo si no existen para que el script se pueda ejecutar
if not os.path.exists(CARPETA_RAIZ_PDFS):
    os.makedirs(os.path.join(CARPETA_RAIZ_PDFS, "subcarpeta"))

# Llamamos a la funci√≥n principal para iniciar la conversi√≥n
procesar_pdfs_recursivamente(CARPETA_RAIZ_PDFS, CARPETA_RAIZ_TEXTOS)
```

## FASE 2 - PROCESAMIENTO DATOS 
- ELIMINAR DUPLICADOS -
Para procesar los TXT se debe confirmar que todos los archivos sean distintos, sin datos duplicados.
El codigo identifica los archivos UNICOS y los mueve a una carpeta raiz manteniendo la estructura.
La salida es una nueva carpeta que mantiene la estructura pero sin archivos duplicados.

```python
import os
import hashlib
import shutil

def calcular_hash(ruta_archivo):
    """Calcula el hash SHA-256 del contenido de un archivo."""
    sha256_hash = hashlib.sha256()
    try:
        with open(ruta_archivo, "rb") as f:
            for byte_block in iter(lambda: f.read(4096), b""):
                sha256_hash.update(byte_block)
        return sha256_hash.hexdigest()
    except IOError:
        print(f"No se pudo leer el archivo: {ruta_archivo}")
        return None

def limpiar_y_copiar_unicos(carpeta_entrada, carpeta_salida):
    """
    Identifica archivos √∫nicos y los copia a una nueva carpeta manteniendo la estructura.
    Reporta los archivos que fueron identificados como duplicados.
    """
    hashes = {}
    duplicados_reporte = {}
    
    print(f"Escaneando archivos en: '{carpeta_entrada}'...")

    # 1. PRIMERA PASADA: Identificar todos los archivos √∫nicos y duplicados
    for dirpath, _, filenames in os.walk(carpeta_entrada):
        for nombre_archivo in filenames:
            if nombre_archivo.lower().endswith(".txt"):
                ruta_completa = os.path.join(dirpath, nombre_archivo)
                file_hash = calcular_hash(ruta_completa)
                
                if file_hash:
                    if file_hash in hashes:
                        # Si el hash ya existe, es un duplicado. Lo guardamos para el reporte.
                        original = hashes[file_hash]
                        if original not in duplicados_reporte:
                            duplicados_reporte[original] = []
                        duplicados_reporte[original].append(ruta_completa)
                    else:
                        # Si es un hash nuevo, guardamos la ruta del archivo como el "original".
                        hashes[file_hash] = ruta_completa

    # 2. SEGUNDA PASADA: Copiar solo los archivos √∫nicos a la nueva carpeta
    print(f"\nCopiando archivos √∫nicos a: '{carpeta_salida}'...")
    if not os.path.exists(carpeta_salida):
        os.makedirs(carpeta_salida)

    for ruta_original in hashes.values():
        # Construir la ruta de destino replicando la estructura de carpetas
        ruta_destino = ruta_original.replace(carpeta_entrada, carpeta_salida, 1)
        
        # Crear el directorio de destino si no existe
        directorio_destino = os.path.dirname(ruta_destino)
        os.makedirs(directorio_destino, exist_ok=True)
        
        # Copiar el archivo
        shutil.copy2(ruta_original, ruta_destino)

    print(f"\nCopia de {len(hashes)} archivos √∫nicos completada.")
    return duplicados_reporte

# --- USO ---

if __name__ == "__main__":
    # Carpeta que contiene los archivos originales (con posibles duplicados)
    CARPETA_A_REVISAR = 'Informes_en_Texto'
    
    # Nueva carpeta donde se guardar√°n solo los archivos √∫nicos
    CARPETA_DESTINO_UNICOS = 'Informes_Unicos'
    
    archivos_duplicados = limpiar_y_copiar_unicos(CARPETA_A_REVISAR, CARPETA_DESTINO_UNICOS)
    
    if not archivos_duplicados:
        print("\n‚úÖ ¬°Excelente! No se encontraron archivos duplicados.")
    else:
        print(f"\nüö® Se encontraron y omitieron los siguientes archivos duplicados:")
        for original, copias in archivos_duplicados.items():
            print("-" * 50)
            print(f"Archivo Original (copiado): {original}")
            for copia in copias:
                print(f"   -> Duplicado (omitido):   {copia}")
        print("-" * 50)
```

## FASE 2 - PROCESAMIENTO DATOS 
- ANONIMIZADO DE DATOS -
El siguiente paso es Anonimizar cada archivo para bloquear los datos sensibles y relacionados con la institucion de donse de obtuvieron.
Se realiza bloqueo de patrones especificos como Normbre, Rut, Apellidos, Fechas y otro bloqueo por bloque, donde se define un marcador inicial y un marcador final.
Como salida se tiene una doble anonimizacion por seguridad donde los archivos se van a una nueva carpeta la cual mantiene la estructura inical.
Esta carpeta es la que se encuentra disponible en el proyecto, al ser repositorios publicos no se pueden subir datos sensibles.

```python
import re
import os

# FUNCIONES DE ANONIMIZACI√ìN

def anonimizar_patrones_especificos(texto):
    """
    Busca y reemplaza patrones de datos sensibles espec√≠ficos dentro de un texto.
    Ideal para el cuerpo del informe.
    """
    # Reemplazar n√∫meros de identificaci√≥n (ej. formato RUT chileno)
    texto = re.sub(r'\b\d{1,2}\.?\d{3}\.?\d{3}-[\dkK]\b', '[ID_ANONIMIZADO]', texto)
    # Reemplazar fechas
    texto = re.sub(r'\b\d{1,2}(/|-|\s+de\s+)\w+(\s+de\s+|-|/)\d{2,4}\b', '[FECHA_ANONIMIZADA]', texto, flags=re.IGNORECASE)
    # Reemplazar nombres despu√©s de etiquetas comunes (ejemplo simple)
    texto = re.sub(r'(Paciente:|Nombre:|Dr\.|Dra\.:)\s*([A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+(\s+[A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+)*)', r'\1 [NOMBRE_ANONIMIZADO]', texto, flags=re.IGNORECASE)
    return texto

def anonimizar_patrones_especificos(texto):
    """
    Busca y reemplaza patrones de datos sensibles espec√≠ficos dentro de un texto.
    Ideal para el cuerpo del informe.
    """
    texto = re.sub(r'\b\d{1,2}\.?\d{3}\.?\d{3}-[\dkK]\b', '[ID_ANONIMIZADO]', texto)
    texto = re.sub(r'\b\d{1,2}(/|-|\s+de\s+)\w+(\s+de\s+|-|/)\d{2,4}\b', '[FECHA_ANONIMIZADA]', texto, flags=re.IGNORECASE)
    texto = re.sub(r'(Paciente:|Nombre:|Dr\.|Dra\.:)\s*([A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+(\s+[A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+)*)', r'\1 [NOMBRE_ANONIMIZADO]', texto, flags=re.IGNORECASE)
    return texto

def anonimizar_informe_por_bloques(texto_completo):
    """
    Anonimiza un informe dividi√©ndolo en encabezado, cuerpo y pie de p√°gina,
    usando marcadores con una l√≥gica de b√∫squeda prioritaria para el inicio.
    """
    # 1. Se definen los marcadores de inicio con prioridad.
    marcador_inicio_primario = r"\bExamen:"
    marcador_inicio_secundario = r"Fecha Examen"
    
    # El marcador de fin no cambia.
    marcador_fin_cuerpo = r"Dr(\(a\))?\.|\bAtentamente\b|Dra(\(a\))?\.|Dra.(\(a\))?\.|Dr.(\(a\))?\."

    # 2. Se busca primero el marcador prioritario "Examen:".
    match_inicio = re.search(marcador_inicio_primario, texto_completo, re.IGNORECASE)

    # 3. Si no se encuentra, se busca el marcador secundario "Fecha Examen".
    if not match_inicio:
        match_inicio = re.search(marcador_inicio_secundario, texto_completo, re.IGNORECASE)

    match_fin = re.search(marcador_fin_cuerpo, texto_completo, re.IGNORECASE)


    if match_inicio and match_fin:
        pos_inicio_cuerpo = match_inicio.start()
        pos_fin_cuerpo = match_fin.start()
        cuerpo = texto_completo[pos_inicio_cuerpo:pos_fin_cuerpo]
        
        encabezado_anonimizado = "[BLOQUE DE DATOS DEL PACIENTE ANONIMIZADO]\n" + "-"*40 + "\n"
        pie_pagina_anonimizado = "\n" + "-"*40 + "\n[BLOQUE DE FIRMA Y DATOS DEL M√âDICO ANONIMIZADO]"
        cuerpo_anonimizado = anonimizar_patrones_especificos(cuerpo)
        
        return encabezado_anonimizado + cuerpo_anonimizado + pie_pagina_anonimizado
    else:
        print(f"  - ADVERTENCIA: Marcadores no encontrados. Se aplicar√° anonimizaci√≥n de patrones a todo el archivo.")
        return anonimizar_patrones_especificos(texto_completo)
        
def anonimizar_estructura_completa(carpeta_raiz_entrada, carpeta_raiz_salida):
    """
    Anonimiza archivos .txt en una carpeta y todas sus subcarpetas,
    replicando la estructura en una carpeta de salida.
    """
    print(f"Iniciando anonimizaci√≥n recursiva desde: '{carpeta_raiz_entrada}'")
    archivos_procesados = 0

    # os.walk() recorre todo el √°rbol de directorios.
    for dirpath, _, filenames in os.walk(carpeta_raiz_entrada):
        
        # 1. Crear la estructura de carpetas correspondiente en la salida.
        ruta_salida_actual = dirpath.replace(carpeta_raiz_entrada, carpeta_raiz_salida, 1)
        if not os.path.exists(ruta_salida_actual):
            os.makedirs(ruta_salida_actual)
            print(f"Creando directorio: {ruta_salida_actual}")

        # 2. Procesar todos los archivos .txt en el directorio actual.
        for nombre_archivo in filenames:
            if nombre_archivo.lower().endswith(".txt"):
                
                ruta_archivo_entrada = os.path.join(dirpath, nombre_archivo)
                ruta_archivo_salida = os.path.join(ruta_salida_actual, nombre_archivo)
                
                print(f"  - Anonimizando: {nombre_archivo}")

                with open(ruta_archivo_entrada, 'r', encoding='utf-8') as f_in:
                    contenido_original = f_in.read()
                
                # Usamos la funci√≥n de anonimizaci√≥n por bloques
                contenido_anonimizado = anonimizar_informe_por_bloques(contenido_original)
                
                with open(ruta_archivo_salida, 'w', encoding='utf-8') as f_out:
                    f_out.write(contenido_anonimizado)
                
                archivos_procesados += 1

    print(f"\n ¬°Proceso completado! Se anonimizaron {archivos_procesados} archivos.")
    print(f"Los resultados se guardaron en la estructura replicada dentro de: '{carpeta_raiz_salida}'")

# ---USO ---

# Carpeta ra√≠z que contiene los archivos de texto originales (con subcarpetas)
CARPETA_RAIZ_ENTRADA = 'Informes_Unicos'

# Carpeta ra√≠z donde se guardar√° la nueva estructura con los archivos anonimizados
CARPETA_RAIZ_SALIDA = 'Informes_Anonimizados'

# Llamamos a la nueva funci√≥n principal para iniciar el proceso
anonimizar_estructura_completa(CARPETA_RAIZ_ENTRADA, CARPETA_RAIZ_SALIDA)
```

## FASE 2 - PROCESAMIENTO DATOS 
- LIMPIEZA DE DATOS -
Se definen las funciones para limpieza de Stopwords, acentos, puntuacion y paso de Mayuscula a minuscula.
Se cargan las mayoria de las librerias para el proyecto para dejar el codigo mas limpio.

La apliacion de estas funciones se realizaran en la creacion del DB 

```python
# 1. PREPARACI√ìN DEL ENTORNO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

from sklearn.metrics import (classification_report, confusion_matrix, precision_recall_curve, 
                             auc, ConfusionMatrixDisplay, roc_curve, roc_auc_score)
import os
import string
import nltk
import unicodedata
import joblib

# Se Descarga y define las 'stopwords'
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('spanish'))

def remover_acentos(texto):
    forma_nfd = unicodedata.normalize('NFD', texto)
    return "".join([c for c in forma_nfd if not unicodedata.combining(c)])

def preprocess_text(text):
    text = text.lower()
    text = remover_acentos(text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)
```

## FASE 3 - CREACION DEL DATASET DESDE CARPETA 
Se crea el Dataet ordenado y se realiza la funcion para etiquetar (Criticos = 0 , No Criticos = 1).
Se aplican las funciones anteriores para obtener texto limpio en el dataset

```python
def crear_dataset_desde_carpetas(ruta_no_criticos, ruta_criticos):
    textos, etiquetas = [], []
    for dirpath, _, filenames in os.walk(ruta_no_criticos):
        for fname in filenames:
            if fname.endswith(".txt"):
                with open(os.path.join(dirpath, fname), 'r', encoding='utf-8') as f:
                    textos.append(f.read()); etiquetas.append(0)
    for dirpath, _, filenames in os.walk(ruta_criticos):
        for fname in filenames:
            if fname.endswith(".txt"):
                with open(os.path.join(dirpath, fname), 'r', encoding='utf-8') as f:
                    textos.append(f.read()); etiquetas.append(1)
    return pd.DataFrame({'texto': textos, 'etiqueta': etiquetas})

CARPETA_NO_CRITICOS = 'Informes_Anonimizados/no_criticos'
CARPETA_CRITICOS = 'Informes_Anonimizados/criticos'
if not os.path.exists(CARPETA_NO_CRITICOS): os.makedirs(CARPETA_NO_CRITICOS)
if not os.path.exists(CARPETA_CRITICOS): os.makedirs(CARPETA_CRITICOS)
# Crear archivos dummy si las carpetas est√°n vac√≠as (EN CASO DE FALLA)
if not os.listdir(CARPETA_NO_CRITICOS):
    with open(os.path.join(CARPETA_NO_CRITICOS, "informe1.txt"), "w") as f: f.write("examen normal sin hallazgos")
if not os.listdir(CARPETA_CRITICOS):
    with open(os.path.join(CARPETA_CRITICOS, "informe2.txt"), "w") as f: f.write("se observa una fractura critica")
        
df = crear_dataset_desde_carpetas(CARPETA_NO_CRITICOS, CARPETA_CRITICOS)
df['texto_limpio'] = df['texto'].apply(preprocess_text)
df.head()
    
```

## FASE 5 - DIVISION, VECTORIZACION, ENTRENAMIENTO Y GUARDADO DE MODELOS
Se decide realizar 2 tipos de modelos
- Regresion Logistica
- Naive Bayes

Se Guardan los modelos para futuras pruebas y comparaciones.

```python
# Division
X = df['texto_limpio']
y = df['etiqueta']

# Vectorizacion
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# ENTRENAMIENTO Y GUARDADO DE MODELOS

joblib.dump((X_test, y_test), 'datos_de_prueba.pkl')
print("\n Datos de prueba (X_test, y_test) guardados en 'datos_de_prueba.pkl'")

# -Modelo 1: Regresi√≥n Log√≠stica
print("\n--- Entrenando y Guardando Modelo: Regresi√≥n Log√≠stica ---")
pipeline_lr = Pipeline([
    ('vectorizer', TfidfVectorizer(ngram_range=(1, 3))),
    ('classifier', LogisticRegression(random_state=42, class_weight='balanced'))
])
pipeline_lr.fit(X_train, y_train)
joblib.dump(pipeline_lr, 'pipeline_regresion_logistica.pkl')
y_pred_lr = pipeline_lr.predict(X_test)
y_pred_probs_lr = pipeline_lr.predict_proba(X_test)[:, 1]
print(" Modelo de Regresi√≥n Log√≠stica guardado como 'pipeline_regresion_logistica.pkl'")

# -Modelo 2: Naive Bayes
print("\n--- Entrenando y Guardando Modelo: Naive Bayes ---")
pipeline_nb = Pipeline([
    ('vectorizer', TfidfVectorizer(ngram_range=(1, 3))),
    ('classifier', MultinomialNB())
])
pipeline_nb.fit(X_train, y_train)
joblib.dump(pipeline_nb, 'pipeline_naive_bayes.pkl')
y_pred_nb = pipeline_nb.predict(X_test)
y_pred_probs_nb = pipeline_nb.predict_proba(X_test)[:, 1]
print(" Modelo de Naive Bayes guardado como 'pipeline_naive_bayes.pkl'")
```

## EVALUACION Y COMPARACION DE METRICAS DE LOS MODELOS

```python
# --- FUNCIONES DE EVALUACION ---

def evaluacion_avanzada(nombre_modelo, y_true, y_pred, y_probs):
    print("\n" + "="*60)
    print(f" EVALUACI√ìN AVANZADA DEL MODELO: {nombre_modelo}")
    print("="*60)
    
    print("\n--- Reporte de Clasificaci√≥n ---")
    print(classification_report(y_true, y_pred, target_names=['No Cr√≠tico (0)', 'Cr√≠tico (1)']))
    
    # 2. Matriz de Confusi√≥n 
    print("\nMostrando Matriz de Confusi√≥n...")
    fig_cm, ax_cm = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred, ax=ax_cm, display_labels=['No Cr√≠tico', 'Cr√≠tico'], cmap='Blues')
    plt.title(f'Matriz de Confusi√≥n - {nombre_modelo}', fontsize=14)
    plt.savefig(f'matriz_confusion_{nombre_modelo.replace(" ", "_").lower()}.png')
    plt.show()
    
    # 3. Curva PR 
    print("\nMostrando Curva Precisi√≥n-Recall...")
    precision, recall, _ = precision_recall_curve(y_true, y_probs)
    pr_auc = auc(recall, precision)
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, lw=2, color='blue', label=f'PR AUC = {pr_auc:.2f}')
    plt.xlabel('Recall'); plt.ylabel('Precisi√≥n'); plt.title(f'Curva PR - {nombre_modelo}'); plt.legend(); plt.grid(True)
    plt.savefig(f'curva_pr_{nombre_modelo.replace(" ", "_").lower()}.png')
    plt.show()

    # 4. CURVA ROC-AUC
    print("\nMostrando Curva ROC-AUC...")
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = roc_auc_score(y_true, y_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--') # L√≠nea de referencia
    plt.xlabel('Tasa de Falsos Positivos (FPR)'); plt.ylabel('Tasa de Verdaderos Positivos (TPR)'); plt.title(f'Curva ROC - {nombre_modelo}'); plt.legend(loc="lower right"); plt.grid(True)
    plt.savefig(f'curva_roc_{nombre_modelo.replace(" ", "_").lower()}.png')
    plt.show()
    
    # 5. Recall@k
    def calculate_recall_at_k(y_true_k, y_scores_k, k):
        desc_score_indices = np.argsort(y_scores_k, kind="mergesort")[::-1]
        top_k_indices = desc_score_indices[:k]
        num_positives = np.sum(y_true_k.iloc[top_k_indices])
        total_positives = np.sum(y_true_k)
        return num_positives / total_positives if total_positives > 0 else 0
        
    print("\n--- M√©trica de Ranking: Recall@k ---")
    total_criticos = np.sum(y_true)
    for k in [10, 20, 30, 50]:
        if k <= len(y_true):
            recall_k = calculate_recall_at_k(y_true, pd.Series(y_probs), k)
            encontrados = int(recall_k * total_criticos)
            print(f"Recall@{k}: {recall_k:.2%} ({encontrados} de {total_criticos} cr√≠ticos encontrados)")

#EVALUACION MODELO REGRESION LOGISTICA
evaluacion_avanzada("Regresi√≥n Log√≠stica", y_test, y_pred_lr, y_pred_probs_lr)

#EVALUACION MODELO NAIVE BAYE
evaluacion_avanzada("Naive Bayes", y_test, y_pred_nb, y_pred_probs_nb)
```

## CONCLUSIONES 
### Conclusi√≥n 1: Rendimiento en General similares para ambos Modelos.

- Curvas ROC y PR: Las formas de las curvas y, lo que es m√°s importante, las √°reas bajo la curva (AUC), son las mismas para ambos.

- PR AUC = 0.93 para ambos modelos. Esto muestra que ambos mantienen una alta calidad en sus predicciones a medida que intentan encontrar m√°s casos.

Esto significa que el potencial predictivo subyacente de ambos algoritmos en tu conjunto de datos es el mismo.

### Conclusi√≥n 2: A pesar de tener el mismo potencial, las matrices de confusi√≥n revelan una diferencia fundamental en su comportamiento con el umbral de decisi√≥n est√°ndar (50%).
- Naive Bayes es m√°s "sensible". Prefiere arriesgarse a generar algunas falsas alarmas adicionales (16 Falsos Positivos) para asegurarse de encontrar la mayor cantidad posible de casos cr√≠ticos (solo omiti√≥ 8).
- Regresi√≥n Log√≠stica es m√°s "conservador". Logra ser un poco m√°s preciso y genera menos falsas alarmas (14 Falsos Positivos), pero lo hace a costa de ser menos sensible y omitir m√°s casos cr√≠ticos (12 en total).

### Conclusion Final :
Para un problema donde el objetivo es detectar informes "cr√≠ticos", el error m√°s costoso es un Falso Negativo (omitir un caso que requer√≠a atenci√≥n urgente).
Naive Bayes es el claro ganador porque comete significativamente menos de este tipo de error peligroso (8 omitidos vs. 12 de la Regresi√≥n Log√≠stica).
Aunque la Regresi√≥n Log√≠stica es ligeramente m√°s precisa, esa peque√±a ganancia en "punter√≠a" no compensa el mayor riesgo de dejar pasar informes importantes.
En resumen, dado el objetivo, es preferible lidiar con un par de falsas alarmas adicionales si eso significa encontrar m√°s casos verdaderamente cr√≠ticos.

## Optimizar el Vectorizador TF-IDF 
Se puede mejorar la calidad de las "features" ignorando palabras que son demasiado raras o demasiado comunes, ya que suelen producir ruido.
Ocuparemos max_df=0.90 ignora palabras que est√°n en m√°s del 90% de los informes (como "informe", "paciente", etc.).


```python
# Pipeline con TF-IDF optimizado
pipeline_nb_optimizado = Pipeline([
    ('vectorizer', TfidfVectorizer(
        ngram_range=(1, 3), 
        max_df=0.90            # Ignorar t√©rminos muy comunes que estan en mas del 90% de los informes
    )),
    ('classifier', MultinomialNB())
])

# Luego entrenas y eval√∫as este nuevo pipeline
pipeline_nb_optimizado.fit(X_train, y_train)
y_pred_nb_opt = pipeline_nb_optimizado.predict(X_test)
y_pred_probs_nb_opt = pipeline_nb_optimizado.predict_proba(X_test)[:, 1]
evaluacion_avanzada("Naive Bayes Optimizado", y_test, y_pred_nb_opt, y_pred_probs_nb_opt)
```

# APLICACION DEL PROYECTO
APLICACI√ìN FINAL - TESTEO DE NUEVOS INFORMES

Este script final integra todo el proyecto. Toma nuevos informes en formato PDF desde una carpeta, aplica la misma cadena de procesamiento con la que fue entrenado el modelo y utiliza nuestro mejor clasificador (`Naive Bayes` con umbral optimizado) para determinar si un informe es "Cr√≠tico" o "No Cr√≠tico", mostrando la confianza de su predicci√≥n.

```python
import fitz  # PyMuPDF
import re
import os
import joblib
import string
import unicodedata
from nltk.corpus import stopwords
import warnings
import shutil

warnings.filterwarnings("ignore", category=FutureWarning, module='nltk')

# ==============================================================================
# FUNCIONES ORIGINALES
# ==============================================================================

def extraer_texto_de_pdf(ruta_pdf):
    """Funci√≥n original de tu proyecto para extraer texto de PDF."""
    try:
        with fitz.open(ruta_pdf) as doc:
            texto_completo = ""
            for pagina in doc:
                bloques = pagina.get_text("blocks", sort=True)
                for b in bloques:
                    texto_completo += b[4]
        return texto_completo
    except Exception as e:
        print(f"  - ERROR al leer el archivo {os.path.basename(ruta_pdf)}: {e}")
        return ""

def anonimizar_patrones_especificos(texto):
    """Funci√≥n original para reemplazar patrones espec√≠ficos."""
    texto = re.sub(r'\\b\\d{1,2}\\.?\\d{3}\\.?\\d{3}-[\\dkK]\\b', '[ID_ANONIMIZADO]', texto)
    texto = re.sub(r'\\b\\d{1,2}(/|-|\\s+de\\s+)\\w+(\\s+de\\s+|-|/)\\d{2,4}\\b', '[FECHA_ANONIMIZADA]', texto, flags=re.IGNORECASE)
    texto = re.sub(r'(Paciente:|Nombre:|Dr\\.|Dra\\.:)\\s*([A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+(\\s+[A-Z√Å√â√ç√ì√ö√ë][a-z√°√©√≠√≥√∫√±]+)*)', r'\\1 [NOMBRE_ANONIMIZADO]', texto, flags=re.IGNORECASE)
    return texto

def anonimizar_informe_por_bloques(texto_completo):
    """
    Funci√≥n de anonimizado por bloques ORIGINAL y ESTRICTA de tu proyecto.
    Aqu√≠ es donde se buscan los marcadores.
    """
    marcador_inicio_primario = r"\\bExamen:"
    marcador_inicio_secundario = r"Fecha Examen"
    marcador_fin_cuerpo = r"Dr(\\(a\\))?\\.|\\bAtentamente\\b|Dra(\\(a\\))?\\.|Dra.(\\(a\\))?\\.|Dr.(\\(a\\))?\\."

    match_inicio = re.search(marcador_inicio_primario, texto_completo, re.IGNORECASE)
    if not match_inicio:
        match_inicio = re.search(marcador_inicio_secundario, texto_completo, re.IGNORECASE)
    
    match_fin = re.search(marcador_fin_cuerpo, texto_completo, re.IGNORECASE)

    if match_inicio and match_fin:
        pos_inicio_cuerpo = match_inicio.start()
        pos_fin_cuerpo = match_fin.start()
        cuerpo = texto_completo[pos_inicio_cuerpo:pos_fin_cuerpo]
        encabezado_anonimizado = "[BLOQUE DE DATOS DEL PACIENTE ANONIMIZADO]\\n" + "-"*40 + "\\n"
        pie_pagina_anonimizado = "\\n" + "-"*40 + "\\n[BLOQUE DE FIRMA Y DATOS DEL M√âDICO ANONIMIZADO]"
        cuerpo_anonimizado = anonimizar_patrones_especificos(cuerpo)
        return encabezado_anonimizado + cuerpo_anonimizado + pie_pagina_anonimizado
    else:
        print(f"  - ADVERTENCIA: Marcadores no encontrados. Se aplicar√° anonimizaci√≥n de patrones a todo el archivo.")
        return anonimizar_patrones_especificos(texto_completo)

def preprocess_text(text):
    """Funci√≥n de limpieza de texto original de tu proyecto."""
    try:
        stop_words = set(stopwords.words('spanish'))
    except LookupError:
        nltk.download('stopwords')
        stop_words = set(stopwords.words('spanish'))
    
    text = text.lower()
    forma_nfd = unicodedata.normalize('NFD', text)
    text = "".join([c for c in forma_nfd if not unicodedata.combining(c)])
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

# ==============================================================================
#  EJECUCI√ìN DEL SCRIPT
# ==============================================================================

if __name__ == "__main__":
    # --- CONFIGURACI√ìN ---
    RUTA_MODELO = 'pipeline_naive_bayes.pkl'
    CARPETA_PDF_PARA_TESTEAR = 'Informes_Para_Testear'
    CARPETA_TXT_TEMPORAL = 'temp_txt_procesados'
    UMBRAL_CRITICIDAD = 0.4

    # Cargar el modelo
    try:
        modelo_cargado = joblib.load(RUTA_MODELO)
        print(f" Modelo '{RUTA_MODELO}' cargado exitosamente.")
    except FileNotFoundError:
        print(f" ERROR: No se encontr√≥ el archivo del modelo '{RUTA_MODELO}'.")
        exit()

    # Fase 1: Procesar PDFs a TXT temporales
    if not os.path.isdir(CARPETA_PDF_PARA_TESTEAR):
        print(f" ERROR: La carpeta de entrada '{CARPETA_PDF_PARA_TESTEAR}' no existe.")
        exit()

    print(f"\\n--- FASE 1: Procesando PDFs con las funciones originales ---")
    os.makedirs(CARPETA_TXT_TEMPORAL, exist_ok=True)
    
    archivos_pdf = [f for f in os.listdir(CARPETA_PDF_PARA_TESTEAR) if f.lower().endswith('.pdf')]
    if not archivos_pdf:
        print(f" No se encontraron archivos PDF para analizar.")
        exit()

    for nombre_pdf in archivos_pdf:
        ruta_pdf = os.path.join(CARPETA_PDF_PARA_TESTEAR, nombre_pdf)
        print(f"  - Procesando: {nombre_pdf}")
        
        texto_crudo = extraer_texto_de_pdf(ruta_pdf)
        texto_procesado = anonimizar_informe_por_bloques(texto_crudo)
        texto_limpio_final = preprocess_text(texto_procesado)
        
        nombre_txt = os.path.splitext(nombre_pdf)[0] + '.txt'
        ruta_txt_temp = os.path.join(CARPETA_TXT_TEMPORAL, nombre_txt)
        
        with open(ruta_txt_temp, 'w', encoding='utf-8') as f:
            f.write(texto_limpio_final)
            
    # Fase 2: Realizar predicciones
    print(f"\\n--- FASE 2: Realizando predicciones ---")
    
    archivos_txt_limpios = os.listdir(CARPETA_TXT_TEMPORAL)
    
    for nombre_txt in archivos_txt_limpios:
        ruta_txt = os.path.join(CARPETA_TXT_TEMPORAL, nombre_txt)
        
        with open(ruta_txt, 'r', encoding='utf-8') as f:
            texto_a_predecir = [f.read()]

        probabilidades = modelo_cargado.predict_proba(texto_a_predecir)[0]
        prob_critico = probabilidades[1]

        if prob_critico >= UMBRAL_CRITICIDAD:
            clasificacion_final = "Cr√≠tico"
            efectividad = prob_critico
        else:
            clasificacion_final = "No Cr√≠tico"
            efectividad = probabilidades[0]
        
        nombre_original = nombre_txt.replace('.txt', '.pdf')
        print("-" * 50)
        print(f"Archivo: {nombre_original}")
        print(f"  Resultado: {clasificacion_final}")
        print(f"  Confianza: {efectividad:.2%}")
        print("-" * 50)

    # Fase 3: Limpieza
    print(f"\\n--- FASE 3: Limpiando archivos temporales ---")
    try:
        shutil.rmtree(CARPETA_TXT_TEMPORAL)
        print(f" Carpeta '{CARPETA_TXT_TEMPORAL}' eliminada correctamente.")
    except OSError as e:
        print(f" ERROR al eliminar la carpeta temporal: {e}")
```
